```python
import numpy as np
from numpy import linalg
import pandas as pd
import matplotlib.pyplot as plt



class normal_equation_model():
  def model(self, x_input):
    df = pd.read_csv('house_prices.csv')

    y_data = df["Price"].values.reshape(-1,1)
    x_data = df[["SqFt","Bedrooms","Bathrooms"]].values

    X_aug = np.concatenate([np.ones((len(x_data), 1)), x_data], axis=1)

    theta = (np.linalg.inv((X_aug.T) @ (X_aug))) @ (X_aug.T) @ (y_data)

    self.x_input = x_input
    x_input = np.array(x_input)
    x_input_aug = np.concatenate([[1], x_input]).reshape(1,-1)

    predict_y = np.dot(x_input_aug,theta)

    print(predict_y) 

    fig, ax = plt.subplots(1,3, figsize=(15,5))
    for i in range(3):
      ax[i].scatter(X_aug[:,i+1], y_data)
      ax[i].scatter(x_input_aug[:,i+1], predict_y)
      ax[i].set_xlabel(f"Feature: {i}")
      ax[i].set_ylabel(f"Price of Houses ($)")


normal_model = normal_equation_model()
normal_model.model([2100,2,2])

    
```
